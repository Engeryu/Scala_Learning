### Installing Java

The first step is to install Java. Scala source code is intended to be compiled to Java bytecode, so that the resulting executable code runs on a Java virtual machine.

<a href="https://www.oracle.com/java/technologies/javase/javase-jdk8-downloads.html" target="_blank">Download the JDK version</a> for your operating system.

**MacOS:** 

Open the dmg file and run the installation accepting all defaults. <a href="https://docs.oracle.com/javase/10/install/installation-jdk-and-jre-macos.htm#JSJIG-GUID-F575EB4A-70D3-4AB4-A20E-DBE95171AB5F" target="_blank">Link to the MacOS documentation</a>

**Linux:** 

```
> sudo apt install openjdk-14-jdk
```

**For both Linux and macOS:**

Now you can test if Java is correctly installed:

### Compiling and Executing Hello World

Unlike Python, Scala needs to be compiled before the program can run. This means the code you write will be converted into commands the java virtual machine (JVM) can understand. 

A Scala Program can be as simple as a single file, but it also can be a much larger project using many files. We will start small with the eternal Hello World...

Create a file names `HelloWorld.scala`

```
object HelloWorld {
    def main(args: Array[String]): Unit = {
        println("Hello, world!")
    }
}
```

Compile the code

```
> scalac HelloWorld.scala
> ls
HelloWorld.class
HelloWorld.scala 
```

You can see that a file named `HelloWorld.class` was created. This is the compiled code that will be executed on the JVM. Now you can execute it.

```
> scala HelloWorld
Hello, world!
```

Congratulations! You created your first Scala program.
